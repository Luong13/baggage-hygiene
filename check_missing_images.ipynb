```python
missing = []
with open('/data/dataset/!annotations.json', encoding='utf-8') as f:
    j = json.load(f)
    for key,value in j['_via_img_metadata'].items():
        if not value['regions']:
            continue
        path = '/data/dataset/' + value['filename']
        my_file = Path(path)
        if os.path.isfile(path):
            continue
        else:
            missing.append(path)
            print(path)
print(len(missing))
```
